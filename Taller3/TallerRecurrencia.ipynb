```python
from sympy import symbols, Rational, simplify
import sys

n = symbols('n')

a_n =  15 *(Rational(1, 3) ** n)

resultado_3c = simplify(a_n.subs(n, 1245123)) # En la ecuación, n valdrá ese número
sys.set_int_max_str_digits(int(resultado_3c)) # Modificar la cantidad de numeros que se pueden mostrar en pantalla
print(resultado_3c)
```
